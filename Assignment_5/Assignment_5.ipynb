```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files

# Step 1 : Read the grayscale input image

uploaded = files.upload()
for fn in uploaded.keys():
    img_path = fn

f = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)

plt.figure(figsize=(5,5))
plt.imshow(f, cmap='gray')
plt.title("Original Image")
plt.axis("off")
plt.show()

# Step 2 : Plot histogram of the original image

hist_f = cv2.calcHist([f], [0], None, [256], [0,256])

plt.figure(figsize=(5,4))
plt.plot(hist_f, color='black')
plt.title("Original Histogram")
plt.xlabel("Pixel Intensity")
plt.ylabel("Frequency")
plt.show()

# Step 3 : Apply Contrast Stretching

r_min = np.min(f)
r_max = np.max(f)

# Apply formula
stretched = ((f - r_min) / (r_max - r_min) * 255).astype(np.uint8)

plt.figure(figsize=(5,5))
plt.imshow(stretched, cmap='gray')
plt.title("Contrast Stretched Image")
plt.axis("off")
plt.show()

# Histogram of Contrast Stretched Image
hist_stretched = cv2.calcHist([stretched], [0], None, [256], [0,256])

plt.figure(figsize=(5,4))
plt.plot(hist_stretched, color='black')
plt.title("Stretched Histogram")
plt.xlabel("Pixel Intensity")
plt.ylabel("Frequency")
plt.show()


# Step 4 : Apply Histogram Equalization

f_eq = cv2.equalizeHist(f)

plt.figure(figsize=(5,5))
plt.imshow(f_eq, cmap='gray')
plt.title("Histogram Equalized Image")
plt.axis("off")
plt.show()

# Histogram of Equalized Image
hist_eq = cv2.calcHist([f_eq], [0], None, [256], [0,256])

plt.figure(figsize=(5,4))
plt.plot(hist_eq, color='black')
plt.title("Equalized Histogram")
plt.xlabel("Pixel Intensity")
plt.ylabel("Frequency")
plt.show()


# Step 5 : Compare results (Horizontal Comparison)

fig, axes = plt.subplots(2, 3, figsize=(15,8))

# Images
axes[0,0].imshow(f, cmap='gray')
axes[0,0].set_title("Original Image")
axes[0,0].axis("off")

axes[0,1].imshow(stretched, cmap='gray')
axes[0,1].set_title("Contrast Stretched")
axes[0,1].axis("off")

axes[0,2].imshow(f_eq, cmap='gray')
axes[0,2].set_title("Histogram Equalized")
axes[0,2].axis("off")

# Histograms
axes[1,0].plot(hist_f, color='black')
axes[1,0].set_title("Original Histogram")

axes[1,1].plot(hist_stretched, color='black')
axes[1,1].set_title("Stretched Histogram")

axes[1,2].plot(hist_eq, color='black')
axes[1,2].set_title("Equalized Histogram")

plt.tight_layout()
plt.show()

# Print Results
print("Contrast Stretching spreads pixel values linearly across 0255 range.")
print("Histogram Equalization redistributes pixel values to achieve a more uniform intensity distribution.")
```
