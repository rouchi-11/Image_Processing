```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset_dir="C:/Users/ROUCHI/Pictures/Screenshots/Screenshot 2025-07-11 103349.png"
```

```python

image = cv2.imread('C:/Users/ROUCHI/Pictures/flower.png', cv2.IMREAD_GRAYSCALE)


plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.axis('off')
plt.show()
```

```python
bright_image = cv2.add(image, 50)  
plt.imshow(bright_image, cmap='gray')
plt.title('Brightness Improved')
plt.axis('off')
plt.show()
```

```python
dark_image = cv2.subtract(image, 50)  
plt.imshow(dark_image, cmap='gray')
plt.title('Brightness Reduced')
plt.axis('off')
plt.show()
```

```python
_, threshold_image = cv2.threshold(image, 127, 255, cv2.THRESH_BINARY)
plt.imshow(threshold_image, cmap='gray')
plt.title('Thresholded Image')
plt.axis('off')
plt.show()
```

```python
negative_image = 255 - image
plt.imshow(negative_image, cmap='gray')
plt.title('Negative Image')
plt.axis('off')
plt.show()
```

```python
# Convert to float and normalize
image_float = image.astype(float)
log_image = np.log1p(image_float)  # log(1 + I)
log_image = (255 / np.log(1 + np.max(image_float))) * log_image  
log_image = np.array(log_image, dtype=np.uint8)

plt.imshow(log_image, cmap='gray')
plt.title('Log Transformed')
plt.axis('off')
plt.show()
```

```python
gamma = 2.2  # Change this value between <1 (brighter) and >1 (darker)
image_normalized = image / 255.0
gamma_corrected = np.power(image_normalized, gamma)
gamma_corrected = np.uint8(gamma_corrected * 255)

plt.imshow(gamma_corrected, cmap='gray')
plt.title(f'Gamma Correction (γ={gamma})')
plt.axis('off')
plt.show()
```
